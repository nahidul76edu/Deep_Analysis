```python
import keras
from keras import backend as K
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
Fatigue_npy = np.load("/content/drive/MyDrive/Mamun_vai/Fatigue500.npy" ,allow_pickle=True)[0:300]
x_train=np.array([i[0] for i in Fatigue_npy])
y_train=np.array([i[1] for i in Fatigue_npy])
x_train=x_train/255

Normal_npy = np.load("/content/drive/MyDrive/Mamun_vai/Normal500.npy" ,allow_pickle=True)[0:300]
x_test=np.array([i[0] for i in Normal_npy])
y_test=np.array([i[1] for i in Normal_npy])
x_test=x_test/255
```

```python
y_train_1=[]
for i in y_train:
  if i[0]==1 and i[1]== 0:
    y_train_1.append(0)

y_test_1=[]
for i in y_test:
  if i[0]==0 and i[1]== 1:
    y_test_1.append(1)
```

```python
X = np.concatenate((x_train, x_test))
Y = np.concatenate((y_train_1, y_test_1))
```

```python
from sklearn.model_selection import train_test_split
# Splitting the dataset into the Training set and Test set
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 0)
```

```python
model = keras.models.load_model('/content/drive/MyDrive/Fatigue_Dataset/new_model')
```

```python
total_layer=len(model.layers)
#############################################################################################################
for k in range(total_layer):
  layer_of_interest=k
  intermediate_tensor_function = K.function([model.layers[0].input],[model.layers[layer_of_interest].output])
  intermediate_tensor = intermediate_tensor_function([x_train[0,:].reshape(-1,224,224,3)])[0]
 
  intermediates = []
  color_intermediates = []
  for i in range(len(x_test)):
    output_class = y_test[i]
    intermediate_tensor = intermediate_tensor_function([x_test[i,:].reshape(-1,224,224,3)])[0]
    p = intermediate_tensor.reshape(1,-1)
    intermediates.append(p[0])

    if(output_class == 0):
        color_intermediates.append("#FF0000")
    elif(output_class == 1):
        color_intermediates.append("#0000FF")

 
  tsne = TSNE(n_components=2, random_state=0)
 
  #x = np.reshape(intermediates, (-1, 224,224,64)).T
  intermediates_tsne = tsne.fit_transform(intermediates)
 
 
  plt.figure(figsize=(4, 4))
  plt.scatter(x = intermediates_tsne[:,0], y=intermediates_tsne[:,1], color=color_intermediates)
  plt.show()
```
